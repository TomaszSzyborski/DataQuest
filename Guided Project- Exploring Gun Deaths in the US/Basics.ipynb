# Exploration of Gun Deaths in US

```python
import csv

with open("guns.csv", "r") as csvfile:
    data = list(csv.reader(csvfile, delimiter=','))
```

```python
data[:5]
```

```python
headers = data[0]
data = data[1:]
```

```python
data[:5]
```

```python
years = [row[1] for row in data]
```

```python
years[:5]
```

```python
from collections import Counter
year_counts = dict(Counter(years))
```

```python
year_counts
```

```python
from datetime import datetime

dates = [datetime(year=int(row[1]), month=int(row[2]), day=1) for row in data]
```

```python
dates[:5]
```

```python
date_counts = dict(Counter(dates))
```

```python
date_counts
```

```python
sexes = [row[headers.index("sex")] for row in data]
sex_counts = dict(Counter(sexes))
```

```python
sex_counts
```

```python
races = [row[headers.index("race")] for row in data]
race_counts = dict(Counter(races))
```

```python
race_counts
```

# Filtering By Intent

```python
intents = [row[3] for row in data]
homicide_race_counts = {}
for i,race in enumerate(races):
    if race not in homicide_race_counts:
        homicide_race_counts[race] = 0
    if intents[i] == "Homicide":
        homicide_race_counts[race] += 1

race_per_hundredk = {}
for k,v in homicide_race_counts.items():
    race_per_hundredk[k] = (v / mapping[k]) * 100000

race_per_hundredk
```
